```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city = pd.read_csv(city_data_to_load)
ride = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
merged = city.merge(ride, on="city")
# Display the data table for preview
merged
```

```python
city.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban = merged.loc[merged["type"] =="Urban"]
urban = urban.groupby("city")
urban_mean = urban.mean()
urban_count =urban.count()
urban_y = urban_mean["fare"]
urban_x = urban_count["driver_count"]
urban_size = urban_mean["driver_count"]*10

rural = merged.loc[merged["type"] =="Rural"]
rural = rural.groupby("city")
rural_mean = rural.mean()
rural_count =rural.count()
rural_y = rural_mean["fare"]
rural_x = rural_count["driver_count"]
rural_size = rural_mean["driver_count"]*10

suburban = merged.loc[merged["type"] =="Suburban"]
suburban = suburban.groupby("city")
suburban_mean = suburban.mean()
suburban_count =suburban.count()
suburban_y = suburban_mean["fare"]
suburban_x = suburban_count["driver_count"]
suburban_size = suburban_mean["driver_count"]*10

# Build the scatter plots for each city types
plt.scatter(urban_x, urban_y, s=urban_size,  color = "lightcoral", alpha=0.8, edgecolors="black", label="Urban")
plt.scatter(suburban_x, suburban_y, s=suburban_size, color = "lightskyblue", alpha=0.8, edgecolors="black", label = "Suburban")
plt.scatter(rural_x, rural_y, s=rural_size, color= "gold", alpha=0.8, edgecolors="black", label="Rural")
# Incorporate the other graph properties
plt.grid()
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare($)")
plt.title("Pyber Ride Sharing Data (2016)")
# Create a legend
plt.legend()
# Incorporate a text label regarding circle size
plt.text(42, 35, "Note:\nCircle size correlates with driver count per city")

# Save Figure
plt.savefig("PyberScatterPlot")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
types = merged.groupby("type")
types_sum = types.sum()
fares = types_sum["fare"]
fares = list(fares)
names = ["Rural", "Suburban", "Urban"]
explode = (0,0,0.1)

# Build Pie Chart
plt.pie(fares, labels = names, explode = explode, colors=["gold", "lightskyblue", "lightcoral"], autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("FaresByCity")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
types = merged.groupby("type")
types = types.count()

rides = types["city"]
rides = list(rides)
names = ["Rural", "Suburban", "Urban"]
explode = (0,0,0.1)

# Build Pie Chart
plt.pie(rides, labels = names, explode = explode, colors=["gold", "lightskyblue", "lightcoral"], autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("RidesByCity")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
drivers = city.groupby("type")
drivers = drivers.sum()
drivers = list(drivers["driver_count"])
drivers
# Build Pie Charts
plt.pie(drivers, labels = names, explode = explode, colors=["gold", "lightskyblue", "lightcoral"], autopct="%1.1f%%", shadow=True, startangle=170)
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("DriversByCity")

```
